```r
##Load Libraries

library(ggplot2)
library(tidyverse)
library(dplyr)
library(readr)
library(cowplot)
library(olsrr)
library(readr) 
library(caret)
library(pscl)
library(lmtest)
library(ipred)
library(survival)
library(ResourceSelection)
library(survey)
library(lmtest)
library(pROC)
library(DescTools)

## NOTE: This is a proof of concept. Further validation work needs to take place.
##need to create train/datasets

data = read.csv('tv_shows.csv', stringsAsFactors = F)
head(data) #allows you to check the data, first few entries 
summary(data) #produce result summaries of the results of various model fitting functions.
dim(data) #the dimension (e.g. the number of columns and rows) of a matrix, array or data frame. 
str(data) 

data$Title <- as.character(data$Title)
data$Hulu <- as.integer(data$Hulu=="1")
str(data)

library(tm)
corpus <- Corpus(VectorSource(data$Title))

corpus <- tm_map(corpus, tolower)
corpus <- tm_map(corpus, removePunctuation)
corpus <- tm_map(corpus, stemDocument)

freq <- DocumentTermMatrix(corpus)
freq

freq <- removeSparseTerms(freq, 0.90)
freq

newcorpus <- as.data.frame(as.matrix(freq))
colnames(newcorpus) <- make.names(colnames(newcorpus))
newcorpus$Hulu<- data$Hulu

library(caTools)
library(e1071)
set.seed(1)
split <- sample.split(newcorpus$Hulu, SplitRatio = 0.7)
train <- subset(newcorpus, split==TRUE)
test <- subset(newcorpus, split==FALSE)
naiveHulu <- naiveBayes(as.factor(Hulu)~., data=train)
predictnaiveHulu <- predict(naiveHulu, newdata = test, type="class")
table(predictnaiveHulu, test$Hulu)
```
